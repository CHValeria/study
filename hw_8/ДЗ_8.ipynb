**Работа с API**

```python
import requests
import pandas as pd
response_books = requests.get('https://www.anapioficeandfire.com/api/books')
books_data = response_books.json()
books_df = pd.DataFrame(books_data)
books_df
```

```python
response_houses = requests.get('https://www.anapioficeandfire.com/api/houses')
houses_data = response_houses.json()
houses_df = pd.DataFrame(houses_data)
houses_df
```

```python
response_houses_with_titles = requests.get('https://www.anapioficeandfire.com/api/houses?hasTitles=true')
houses_with_titles_data = response_houses_with_titles.json()
houses_with_titles_df = pd.DataFrame(houses_with_titles_data)
houses_with_titles_df
```
